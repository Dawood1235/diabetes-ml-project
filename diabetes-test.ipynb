```python
X = diab_dtst.drop(columns = "Outcome", axis = 1)
Y = diab_dtst["Outcome"]
```

```python
print(X)
```

```python
print(Y)
```

```python
scalar = StandardScaler()
```

```python
scalar.fit(X)
```

```python
standard_dt = scalar.transform(X)
```

```python
print(standard_dt)
```

```python
X = standard_dt
```

```python
Y = diab_dtst["Outcome"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel = "linear")
```

```python
classifier.fit(X_train,Y_train)
```

```python
X_train_predict = classifier.predict(X_train)
train_dt_acc = accuracy_score(X_train_predict, Y_train)
```

```python
print(f"The accuracy of training data: {train_dt_acc}")
```

```python
X_test_predict = classifier.predict(X_test)
test_dt_acc = accuracy_score(X_test_predict, Y_test)
```

```python
print(f"The accuracy of test data: {test_dt_acc}")
```
